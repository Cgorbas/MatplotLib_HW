```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import csv
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# File to Load (Remember to Change These)

mouse_file=os.path.join("data","mouse_drug_data.csv")
trial_file=os.path.join("data","clinicaltrial_data.csv")

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df=pd.read_csv(mouse_file)
trial_df=pd.read_csv(trial_file)


#merge data files
#clinical_df = pd.merge(trial_df,mouse_df, on="Mouse ID")
clinicaldf = pd.merge(mouse_df,trial_df, on="Mouse ID")

#clinicaldf.head(5)
cols=clinicaldf.columns.tolist()
cols.insert(4, cols.pop(cols.index('Mouse ID')))

clinicaldf=clinicaldf.reindex(columns=cols)
clinicaldf['DrugGrp']=clinicaldf['Drug']
clinicaldf['Metastatic Sites']=clinicaldf['Metastatic Sites'].astype(float)

clinicaldf.sort_values(['DrugGrp'],ascending=[True])

clinical_df = clinicaldf.groupby('DrugGrp')

#clinical_df.assign(index=clinical_df.groupby('Drug','Timepoint')).pivot('Timepoint','Drug','val')

clinicaldf.dtypes
clinical_df.head()
```

```python
## Tumor Response to Treatment
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

#tumorVolMn=clinical_df["Tumor Volume (mm3)"].mean()
tumorVolMn=clinical_df["Tumor Volume (mm3)"].mean()
tumorVolstErr=clinical_df["Tumor Volume (mm3)"].sem()
tumorVolstdev=clinical_df["Tumor Volume (mm3)"].std()
tumorVolDrug=clinical_df["DrugGrp"]

# Convert to DataFrame
tumorVol_df = pd.DataFrame({
                            "Tumor Vol Mean":tumorVolMn, 
                            "Tumor Vol StdErr":tumorVolstErr,
                            "Tumor Vol Std Dev":tumorVolstdev
                                                    })
#print(timePoint)
# Preview DataFrame
tumorvold=tumorVol_df.T
tumorVoldf=pd.DataFrame(data=tumorvold)
tumorVoldf.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames

drugTimePiv=clinicaldf.pivot_table(index='Timepoint', columns='Drug', 
                                values='Tumor Volume (mm3)')

drugTimePiv.head(1)
```

```python
# Generate the Plot (with Error Bars)


#define axes
#plot standard error of tumor volume
#tumorVoldf
#Axes
x_axis=np.array(drugTimePiv.index)
#y_axis=np.array(drugTimePiv)
y1=drugTimePiv['Capomulin']
y2=drugTimePiv['Ceftamin']
y3=drugTimePiv['Infubinol']
y4=drugTimePiv['Ketapril']
y5=drugTimePiv['Naftisol']
y6=drugTimePiv['Placebo']
y7=drugTimePiv['Propriva']
y8=drugTimePiv['Ramicane']
y9=drugTimePiv['Stelasyn']
y10=drugTimePiv['Zoniferol']

#Grid Formatting
tick_locations=[value +5 for value in x_axis]
plt.xticks=(tick_locations)
plt.grid(which='major', axis='both')

plt.title("Drug Effectiveness on Tumor Volume")
plt.xlabel("Timepoint")
plt.ylabel("Tumour Volume (mm3)")


#plot
plt.plot(x_axis,y1,color='red', linestyle='solid', marker='o',label='Capomulin')
plt.plot(x_axis,y2,color='red', linestyle='dashed', marker='p',label='Ceftamin')
plt.plot(x_axis,y3,marker='s', linestyle='solid', color='green',label='Infubinol')
plt.plot(x_axis,y4,marker='h', linestyle='dashed', color='black',label='Ketapril')
plt.plot(x_axis,y5,marker='o', linestyle='dashed', color='lightblue',label='Naftisol')
plt.plot(x_axis,y6,marker='+', linestyle='solid', color='orange',label='Placebo')
plt.plot(x_axis,y7,marker='*', linestyle='dashed', color='purple',label='Propriva')
plt.plot(x_axis,y8,marker='<', linestyle='solid', color='brown',label='Ramicane')
plt.plot(x_axis,y9,marker='>', linestyle='dashed', color='grey',label='Stelasyn')
plt.plot(x_axis,y10,marker='x', linestyle='solid', color='darkgreen',label='Zoniferol')
plt.legend(loc="upper left")

# Save the Figure
plt.savefig("data/drugeffect_data.png")
plt.tight_layout()
plot.show()
```
